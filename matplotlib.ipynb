```python
import pandas as pd
import matplotlib.pyplot as pld
```

```python
df=pd.read_csv("pizza.csv")
df.head(10)
```

```python
df.dropna(inplace=True)
df.head()
```

```python
pld.plot(df["name"].count(),df["size"].count())
pld.show
```

```python
pld.figure(figsize=[40,40])
pld.bar(df["name"].value_counts().index , df["name"].value_counts().values)
pld.show()
```

```python
pld.figure(figsize=[10,10])
pld.barh(df["name"].value_counts().index , df["name"].value_counts().values,color=('orange','green',"y"),label="pizza 1")
pld.title("This is a bar graph")
pld.xlabel("count of pizza")
pld.ylabel("Pizza ka prakar")
pld.grid()
pld.legend(["pizza 1",])
pld.show()
```
